### %timeit tells how long it takes for a `loop` to run

````python
def loop(n):
    k=[]
    for i in range(n):
        k.append(i**2) 
%timeit loop(10)
````

> %% timeit tells how long does it take for a cell to run

````python
%%timeit
def loop(n):
    k=[]
    for i in range(n):
        k.append(i**2) 
loop(10)
````
